```python
import matplotlib.pyplot as plt
import matplotlib.ticker as mticker
import warnings
import csv
```

## Import and setting up our data

I start by bringing in the CSV data into Python program.
I then parse the data using the CSV package.
Lastly, I close the CSV file after we have read the information

```python
with open('./data/USvideos.csv', 'r', encoding="utf8") as datafile:
    youtube_data= csv.DictReader(datafile)
    data_list= []
    for row in youtube_data:
        data_list.append(row)
```

## Format the data before we generate a visualization

For trending YouTube videos in the US, I sort data by the amount of views, compared to the difference of their likes and dislikes (ratio)

```python
views_list= []
def view_function():
    for view_data in data_list:
        views= int(view_data['views'])
        views_list.append(views)
#     print(views_list)
view_function()
```

```python
ratio_list= []
def ratio_function():
    for ratio_data in data_list:
        likes= int(ratio_data['likes'])
        dislikes= int(ratio_data['dislikes'])
        ratio= likes- dislikes
        ratio_list.append(ratio)
#     print(ratio_list)
ratio_function()
```

## Create the visualization

Now that the data is parsed, I use matplotlib to create a graph

```python
# X-axis Values
x= ratio_list


# Y-axis Values
y= views_list


# Error Handeling
def fxn():
    warnings.warn("deprecated", DeprecationWarning)
with warnings.catch_warnings():
    warnings.simplefilter("ignore")
    fxn()

# Initializing
fig, ax= plt.subplots(figsize= (15, 12)) 
ax.scatter(x, y)

# Labeling
ax.set_xticklabels(('0', '0.5 million', '1 million', '1.5 million', '2 million', '2.5 million', '3 million', '3.5 million', '4 million', '4.5 million', '5 million', '5.5 million'))
ax.set_yticklabels(('0', '20 million', '40 million', '60 million', '80 million', '100 million', '120 million', '140 million', '160 million', '180 million', '200 million', '220 million', '240 million'))
ax.set_title("Relationship Between Views and Like/Dislike Ratio")
ax.set_xlabel("Like/Dislike Ratio (Likes minus Dislikes)")
ax.set_ylabel("Views")

# Formating
ax.set_xlim(0, 5500000)
ax.set_ylim(0, 240000000)
ax.locator_params(axis='y', nbins= 20)
ax.locator_params(axis='x', nbins= 20)

# Finalizing
plt.show()
```
